```python
df = pd.read_csv("pppub20.csv")
```

```python
"""
A_AGE = age
A_FNLWGT = fnlwgt
A_SEX = gender
A_MARITL = marital-status
A_PFREL = relationship
A_HGA = education
PRDTRACE = race
PTOT_R (income recode)= income 
A_CLSWKR = class of worker
A_MJOCC = occupation
HRSWK = hours per week
PENATVTY(coded) = native-country
CAP_VAL = capital gain
AGI = adjusted gross income - mainly used to select which rows to keep, nothing else
"""
adult_data_col = ['A_AGE','A_CLSWKR','A_FNLWGT','A_HGA','A_MARITL','A_MJOCC','A_PFREL','PRDTRACE','A_SEX','CAP_VAL','HRSWK','PENATVTY','PTOT_R','AGI']
clean_data = df[adult_data_col]
display(clean_data.head(10))
clean_data.shape
```

```python
#drop all age 16 and under
over16 =  clean_data[clean_data["A_AGE"]>16]
over16.shape
```

```python
#only keep rows with AGI>100, FNLWGT>1, and HRSWK>0 like the 1994 dataset
over16 = over16[over16["AGI"]>100]
over16 = over16[over16["A_FNLWGT"]>1]
over16 = over16[over16["HRSWK"]>0]
over16.shape
```

```python
#df["column1"].replace({"a": "x", "b": "y"}, inplace=True)
#convert workclass code to string, similar to adult dataset
over16["A_CLSWKR"].replace({0:"?",1:"Private",2:"Federal-gov",3:"State-gov",4:"Local-gov",5:"Self-emp-inc",6:"Self-emp-not-inc",7:"Without-pay",8:"Never-worked"},inplace=True)
over16.head(10)
```

```python
#add educational-num column
edu_num_added = over16.copy()
edu_num_added = edu_num_added[edu_num_added["A_HGA"]!=0]
edu_num_added["educational-num"] = edu_num_added["A_HGA"]
edu_num_added["educational-num"].replace({0:"?",31:1,32:2,33:3,34:4,35:5,36:6,37:7,38:8,39:9,40:10,41:11,42:12,43:13,44:14,45:15,46:16},inplace=True)
cols = edu_num_added.columns.tolist()
rearranged_cols = cols[:4] + cols[-1:] + cols[4:13]
edu_num_added = edu_num_added[rearranged_cols]
```

```python
#renaming columns to match adult dataset
renamed_columns = edu_num_added.rename(columns={"A_AGE":"age","A_CLSWKR":"workclass","A_FNLWGT":"fnlwgt","A_HGA":"education","A_MARITL":"marital-status","A_MJOCC":"occupation","A_PFREL":"relationship","PRDTRACE":"race","A_SEX":"gender","CAP_VAL":"capital-gain","HRSWK":"hours-per-week","PENATVTY":"native-country","PTOT_R":"income"})
```

```python
#recoding education column
renamed_columns["education"].replace({0:"?",31:"Preschool",32:"1st-4th",33:"5th-6th",34:"7th-8th",35:"9th",36:"10th",37:"11th",38:"12th",39:"HS-grad",40:"Some-college",41:"Assoc-voc",42:"Assoc-acdm",43:"Bachelors",44:"Masters",45:"Prof-school",46:"Doctorate"},inplace=True)
renamed_columns
```

```python
renamed_columns.to_csv("2020_ASEC_data_cleaninginprogress.csv",index=False)
```

```python
#recoding marital-status
renamed_columns["marital-status"].replace({1:"Married-civ-spouse",2:"Married-AF-spouse",3:"Married-spouse-absent",4:"Widowed",5:"Divorced",6:"Separated",7:"Never-married"},inplace=True)
renamed_columns
```

```python
#recoding occupation
renamed_columns["occupation"].replace({0:"?",1:"Exec-managerial",2:"Prof-specialty",3:"Other-service",4:"Sales",5:"Adm-clerical",6:"Farming-fishing",7:"Construction",8:"Craft-repair",9:"Production",10:"Transport-moving",11:"Armed-Forces"},inplace=True)
renamed_columns
```

```python
#recoding relationship
renamed_columns["relationship"].replace({0:"Not-in-family",1:"Husband",2:"Wife",3:"Own-child",4:"Other-relative",5:"Unmarried"},inplace=True)
renamed_columns
```

```python
#recoding race
renamed_columns["race"].replace({1:"White",2:"Black",3:"Amer-Indian-Eskimo",4:"Asian-Pac-Islander",5 :'Other' ,6 :'Other' ,7 :'Other' ,8 :'Other' ,9 :'Other' ,10 :'Other' ,11 :'Other' ,12 :'Other' ,13 :'Other' ,14 :'Other' ,15 :'Other' ,16 :'Other' ,17 :'Other' ,18 :'Other' ,19 :'Other' ,20 :'Other' ,21 :'Other' ,22 :'Other' ,23 :'Other' ,24 :'Other' ,25 :'Other' ,26 :'Other'},inplace=True)
renamed_columns
```

```python
#recoding gender
renamed_columns["gender"].replace({1:"Male",2:"Female"},inplace=True)
renamed_columns
```

```python
#recoding income
renamed_columns2 = renamed_columns.rename(columns={"income":"coded-income"})
```
