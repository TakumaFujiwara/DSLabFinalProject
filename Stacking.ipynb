```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt
import numpy as np

from sklearn.ensemble import StackingClassifier

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
import xgboost as xgb
from catboost import CatBoostClassifier
```

```python
from sklearn.model_selection import cross_val_score
def auc_cv(model,my_X,my_y):
    score = cross_val_score(model, my_X, my_y, scoring="roc_auc", cv = 20)
    return(score)
```

```python
#for 1994
X = pd.read_csv("X_1994.csv")
X.drop(columns=X.columns[0], 
        axis=1, 
        inplace=True)
X.drop(columns=X.columns[4], 
        axis=1, 
        inplace=True)
X = pd.get_dummies(data=X,drop_first=True)
y = pd.read_csv("y_1994.csv")
y.drop(columns=y.columns[0], 
        axis=1, 
        inplace=True)
y = pd.get_dummies(data=y,drop_first=True)

level0 = list()
level0.append(('lore',LogisticRegression(C= 0.1, penalty= 'l2', solver= 'newton-cg')))
level0.append(('rf', RandomForestClassifier(n_estimators = 1200,max_depth = 15,min_samples_split = 5,min_samples_leaf = 1)))
level0.append(('bt', GradientBoostingClassifier(subsample = 1.0, n_estimators = 200, min_samples_split = 10, min_samples_leaf= 5, max_depth= 7, loss= 'exponential', learning_rate= 0.1)))
level0.append(('mlp', MLPClassifier(activation = 'identity' ,solver= 'adam' ,alpha= 0.0001, learning_rate='constant', max_iter = 2000)))
level0.append(('cb',CatBoostClassifier()))
level0.append(('xb',xgb.XGBClassifier()))

level1 = LogisticRegression()

model = StackingClassifier(estimators=level0, final_estimator=level1,cv=20,verbose=2)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
model.fit(X_train, y_train)
preds = model.predict_proba(X_test)
rocScore = roc_auc_score(y_test, preds[:,1])
print("ROC score is:",rocScore)

#score = auc_cv(model,X,y)
#print("Average is:", np.mean(score))
#print(score)
```

```python
#for 2020

X = pd.read_csv("X_2020.csv")
X.drop(columns=X.columns[0], 
        axis=1, 
        inplace=True)

X = pd.get_dummies(data=X,drop_first=True)
y = pd.read_csv("y_2020.csv")
y.drop(columns=y.columns[0], 
        axis=1, 
        inplace=True)
y = pd.get_dummies(data=y,drop_first=True)


level0 = list()
level0.append(('lore',LogisticRegression(C= 0.1, penalty= 'l2', solver= 'newton-cg')))
level0.append(('rf', RandomForestClassifier(n_estimators = 1200,max_depth = 15,min_samples_split = 5,min_samples_leaf = 1)))
level0.append(('bt', GradientBoostingClassifier(subsample = 1.0, n_estimators = 200, min_samples_split = 10, min_samples_leaf= 5, max_depth= 7, loss= 'exponential', learning_rate= 0.1)))
level0.append(('mlp', MLPClassifier(activation = 'identity' ,solver= 'adam' ,alpha= 0.0001, learning_rate='constant', max_iter = 2000)))
level0.append(('cb',CatBoostClassifier(od_wait= 300, iterations= 1000, l2_leaf_reg= 0.1,verbose=False)))
level0.append(('xb',xgb.XGBClassifier(max_depth= 4, min_split_loss= 0, n_estimators= 500, objective= "binary:logistic", subsample= 1,verbose_eval=False)))

level1 = LogisticRegression()

model = StackingClassifier(estimators=level0, final_estimator=level1,cv=20,verbose=2)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
model.fit(X_train, y_train)
preds = model.predict_proba(X_test)
rocScore = roc_auc_score(y_test, preds[:,1])
print("ROC score is:",rocScore)
```
